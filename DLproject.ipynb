```python
from keras.utils import to_categorical

if y_train.ndim == 1 or y_train.shape[1] == 1:
    y_train = to_categorical(y_train, num_classes=3)
if y_test.ndim == 1 or y_test.shape[1] == 1:
    y_test = to_categorical(y_test, num_classes=3)
```

```python
from keras.optimizers import SGD, RMSprop, Adam, Adadelta, Adagrad, Adamax
def create_model():
  model= Sequential()
  #input layer
  model.add(Dense(8,activation='relu',input_dim = X.shape[1]))
  #hidden layers
  model.add(Dense(16,activation='relu'))
  model.add(Dense(8,activation='relu'))
  model.add(Dense(16,activation='relu'))
  #output layer
  model.add(Dense(3,activation='softmax'))
  return model
#compile the model
optimizers = {
    'SGD': SGD(),
    'RMSprop': RMSprop(),
    'Adam': Adam(),
    'Adadelta': Adadelta(),
    'Adagrad': Adagrad(),
    'Adamax': Adamax()
}
for name, optimizer in optimizers.items():
    print(f"Training with {name} optimizer")

    # Create a new model instance
    model = create_model()

    # Compile the model with the current optimizer
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model
    model.fit(X_train, y_train, epochs=10, validation_split=0.3, verbose=0)

    # Evaluate the model
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")
    print("-" * 40)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, RMSprop, Adam, Adadelta, Adagrad, Adamax

def create_model():
    model = Sequential()
    # Input layer
    model.add(Dense(8, activation='relu', input_dim=X_train.shape[1]))
    # Hidden layers
    model.add(Dense(16, activation='relu'))
    model.add(Dense(8, activation='relu'))
    # Output layer
    model.add(Dense(3, activation='softmax'))
    return model

# Dictionary of optimizers
optimizers = {
    'SGD': SGD(),
    'RMSprop': RMSprop(),
    'Adam': Adam(),
    'Adadelta': Adadelta(),
    'Adagrad': Adagrad(),
    'Adamax': Adamax()
}

# Store the accuracy for each optimizer
results = {}

for name, optimizer in optimizers.items():
    print(f"Training with {name} optimizer")

    # Create a new model instance
    model = create_model()

    # Compile the model with the current optimizer
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

    # Train the model and capture the history
    history = model.fit(X_train, y_train, epochs=11, validation_split=0.3, verbose=0)

    # Evaluate the model
    loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
    results[name] = (history, accuracy)
    print(f"Test Loss: {loss}, Test Accuracy: {accuracy}")
    print("-" * 40)

# Plot training and validation accuracy for each optimizer
plt.figure(figsize=(14, 7))
for name, (history, accuracy) in results.items():
    plt.plot(history.history['accuracy'], label=f'{name} Training Accuracy')
    plt.plot(history.history['val_accuracy'], label=f'{name} Validation Accuracy')
plt.title('Training and Validation Accuracy for Different Optimizers')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot test accuracy for each optimizer
optim_names = list(results.keys())
test_accuracies = [acc for _, acc in results.values()]

plt.figure(figsize=(10, 5))
plt.bar(optim_names, test_accuracies)
plt.title('Test Accuracy for Different Optimizers')
plt.xlabel('Optimizer')
plt.ylabel('Test Accuracy')
plt.show()
```
